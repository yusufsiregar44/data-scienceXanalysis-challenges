```python
# project configurations

import sqlalchemy as sa
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import datetime
import os
import zipfile
import uuid
import matplotlib
import seaborn as sns
from os.path import basename
from datetime import datetime, date, time, timedelta


client_pg = sa.create_engine('postgresql://localhost:5432/postgres')
```

```python
# create tables

create_table_transactions = """
CREATE TABLE transactions(
    index INT,
    id uuid PRIMARY KEY,
    currency CHAR(4) NOT NULL,
    amount BIGINT NOT NULL,
    state VARCHAR(30) NOT NULL,
    created_date TIMESTAMP NOT NULL,
    merchant_category VARCHAR(100),
    merchant_country VARCHAR(100),
    entry_method VARCHAR(4) NOT NULL,
    user_id UUID NOT NULL,
    type VARCHAR(20) NOT NULL,
    source VARCHAR(20) NOT NULL
);
"""
create_table_users = """
CREATE TABLE users(
    index INT,
    id uuid PRIMARY KEY,
    has_email INT NOT NULL,
    phone_country VARCHAR(300),
    terms_version DATE,
    created_date TIMESTAMP NOT NULL,
    state VARCHAR(25) NOT NULL,
    country VARCHAR(2),
    birth_year INT,
    kyc VARCHAR(20),
    failed_sign_in_attempts INT
);
"""
create_table_fx_rates = """
CREATE TABLE fx_rates(
    base_ccy VARCHAR(3),
    ccy VARCHAR(10),
    rate DOUBLE PRECISION
);
"""
create_table_currency_details = """
CREATE TABLE currency_details(
    currency VARCHAR(10) PRIMARY KEY,
    iso_code INT,
    exponent INT,
    is_crypto BOOLEAN NOT NULL
);
"""

create_table_countries = """
CREATE TABLE countries(
    code VARCHAR(2),
    name VARCHAR(50),
    code3 VARCHAR(3),
    numcode INT,
    phonecode INT
);
"""

create_table_fraudsters = """
    CREATE TABLE fraudsters(
        user_id UUID NOT NULL,
        index INT
    );
"""

# client_pg.execute(create_table_transactions)
# client_pg.execute(create_table_users)
# client_pg.execute(create_table_fx_rates)
# client_pg.execute(create_table_currency_details)
# client_pg.execute(create_table_countries)
# client_pg.execute(create_table_fraudsters)
```

```python
# drop tables

drop_table_transactions = """DROP TABLE transactions"""
drop_table_users = """DROP TABLE users"""
drop_table_fx_rates = """DROP TABLE fx_rates"""
drop_table_currency_details = """DROP TABLE currrency_details"""
drop_table_countries = """DROP TABLE countries"""
drop_table_fraudsters = """DROP TABLE fraudsters"""

client_pg.execute(drop_table_transactions)
client_pg.execute(drop_table_users)
client_pg.execute(drop_table_fx_rates)
client_pg.execute(drop_table_currrency_details)
client_pg.execute(drop_table_countries)
client_pg.execute(drop_table_fraudsters)
```
